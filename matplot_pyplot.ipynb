```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
year = [2014, 2015, 2016, 2017, 2018]
apl_price = [93.95, 112.15, 104.05, 144.85, 169.49] # apple stock
ms_price = [39.01, 50.29, 57.05, 69.98, 94.39] # microsoft stock
```

```python
# plt.plot(year, apl_price) # plot as a line chart
# plt.scatter(year, ms_price) #plot as a scatter chart
#instead of above two lines, we can also do , 
# plt.plot(year, apl_price,
#         year, ms_price)
# to add colors, we can say
plt.plot(year, apl_price, 'k') ## k means black with
plt.plot(year, ms_price, 'g') ## g = green
plt.xlabel("Year")
plt.ylabel("Stock price")

#when we want to set values on the list i.e min and max value
plt.axis([2013, 2019, 20, 180])
# first two values are min, max for x-axis
# next two values are min, max for y-axis

plt.show()
```
