```python
from matplotlib import pyplot as plt
%matplotlib inline

# Reference video = https://www.youtube.com/watch?v=qErBw-R2Ybk
```

```python
year = [2014, 2015, 2016, 2017, 2018]
apl_price = [93.95, 112.15, 104.05, 144.85, 169.49] # apple stock
ms_price = [39.01, 50.29, 57.05, 69.98, 94.39] # microsoft stock
```

```python
# plt.plot(year, apl_price) # plot as a line chart
# plt.scatter(year, ms_price) #plot as a scatter chart
#instead of above two lines, we can also do , 
# plt.plot(year, apl_price,
#         year, ms_price)
# to add colors, we can say
plt.plot(year, apl_price, 'k') ## k means black with
plt.plot(year, ms_price, 'g') ## g = green
plt.xlabel("Year")
plt.ylabel("Stock price")

#when we want to set values on the list i.e min and max value
plt.axis([2013, 2019, 20, 180])
# first two values are min, max for x-axis
# next two values are min, max for y-axis

plt.show()
```

```python
fig_1 = plt.figure(1, figsize=(6.4, 4.8)) #first number is Sno, 2nd is fig size=(width, height)

# we need to add charts to this figure
chart_1 = fig_1.add_subplot(121) # 1 row with 2 columns and position chart1 at 1st column --> 121
chart_2 = fig_1.add_subplot(122) # 1 row with 2 columns and position chart2 at 2nd column --> 121

# chart_1 = fig_1.add_subplot(224)
# chart_2 = fig_1.add_subplot(221)

chart_1.plot(year, apl_price)
chart_2.scatter(year, ms_price)

plt.show()
```

```python
# To create subplots faster
fig_2, axes = plt.subplots(1, 2, figsize=(20, 5))
axes[0].scatter(year, apl_price)
axes[1].plot(year, ms_price)
plt.show()
```

```python
# To create subplots faster
fig_2, axes = plt.subplots(2, 2, figsize=(20, 5))
axes[0,1].scatter(year, apl_price)
axes[1,0].plot(year, ms_price)
plt.show()
```
